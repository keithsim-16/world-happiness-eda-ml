### Insight 1. Does a population with highly educated females lead to lower birth rates?
> Extract required variables from 'Data' dataframe for analysis  
> - `Education: Primary gross enrol. ratio (female per 100 pop.)`   
> - `Education: Secondary gross enrol. ratio (female per 100 pop.)`  
> - `Education: Tertiary gross enrol. ratio (female per 100 pop.)`  
> - `Fertility rate, total (live births per woman)`

```python
EduLvl_BirthRate_Data = pd.DataFrame(Data[['Education: Primary gross enrol. ratio (female per 100 pop.)', 
                                           'Education: Secondary gross enrol. ratio (female per 100 pop.)', 
                                           'Education: Tertiary gross enrol. ratio (female per 100 pop.)', 
                                           'Fertility rate, total (live births per woman)']])
EduLvl_BirthRate_Data.head()
```

### Check variables independently

```python
EduLvl_BirthRate_Data.describe()
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(4, 3, figsize=(18, 20))

count = 0
for var in EduLvl_BirthRate_Data:
    sb.boxplot(data = EduLvl_BirthRate_Data[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = EduLvl_BirthRate_Data[var], ax = axes[count,1])
    sb.violinplot(data = EduLvl_BirthRate_Data[var], orient = "h", ax = axes[count,2])
    count += 1
```

```python
# Calculate the quartiles
Q1 = EduLvl_BirthRate_Data.quantile(0.25)
Q3 = EduLvl_BirthRate_Data.quantile(0.75)

# Rule to identify outliers
rule = ((EduLvl_BirthRate_Data < (Q1 - 1.5 * (Q3 - Q1))) | (EduLvl_BirthRate_Data > (Q3 + 1.5 * (Q3 - Q1))))

# Count the number of outliers
rule.sum()
```

```python
EduLvl_BirthRate_Data.skew()
```

By observation and computation, we can tell that `Education: Primary gross enrol. ratio (female per 100 pop.)` has the most number of outliers compared to the other variables while `Fertility rate, total (live births per woman)` is moderately skewed at 1.042320.

### Check relationship between education levels and fertility rate

```python
# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(12, 12))
sb.heatmap(EduLvl_BirthRate_Data.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = EduLvl_BirthRate_Data, height=5)
```

```python
cols = list(EduLvl_BirthRate_Data.columns)[:-1]
# Draw the distributions of all variables
count = 0
for item in cols:
    g = sb.jointplot(x = item, 
                 y = "Fertility rate, total (live births per woman)", 
                 data = EduLvl_BirthRate_Data,
                 height = 12,
                 ax = axes[0], 
                 kind = "reg")
    j = sb.regplot(x=item, y="Fertility rate, total (live births per woman)", 
               data=EduLvl_BirthRate_Data, ax = g.ax_joint)
    j.set(ylim=(0, None))
    count += 1
```

- Compared to secondary and tertiary education level, primary education level has the lowest correlation value to fertility rate.  

- It is difficult to derive any conclusion from the primary education level joint plot as most data points are centralised at about 100 (female per 100 in the population) and 1 - 3 fertility rate.  

- However, we can tell that between the three education levels and the fertility rate, all three are negatively correlated. Especially in secondary and tertiary education levels, it is clear that as education enrolment ratio increases, the fertility rate decreases.  

- This allows us to make an observation that birth rate decreases as the population gets more educated, but, this is not conclusive as there may be other factors affecting the fertility rate.

`Education: Primary gross enrol. ratio (male per 100 pop.)` and `Education: Tertiary gross enrol. ratio (male per 100 pop.)` are moderately skewed and has the highest skew values compared to the others at 0.936776 and 0.880256 respectively.

### Insight 3. Happiness Score Compared to Locality
> Extract required variables from 'Data' dataframe for analysis  
> - `country`   
> - `happiness.score`  
> - `latitude`  
> - `longitude`

```python
Map_Data = pd.DataFrame(Data[['country', 'happiness.score', 'latitude', 'longitude']])

Map_Data
```

```python
Map_Data.info()
```

```python
import geopandas as gpd

# From GeoPandas, our world map data
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Creating axes and plotting world map
fig, ax = plt.subplots(figsize=(20, 10))
worldmap.plot(color="lightgrey", ax=ax)

# Plotting our Impact Energy data with a color map
x = Map_Data['longitude']
y = Map_Data['latitude']
z = Map_Data['happiness.score']
plt.scatter(x, y, s=20*z, c=z, alpha=0.6, cmap='RdBu')
plt.colorbar(label='happiness.score')

# Creating axis limits and title
plt.xlim([-180, 180])
plt.ylim([-90, 90])

plt.title("World Happiness Report\n")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

We can observe that the regions that experience the lowest happiness scores are located primarily in Africa, the Middle East and South Asia.
