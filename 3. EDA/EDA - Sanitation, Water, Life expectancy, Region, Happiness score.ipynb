## Sanitation and Drinking Water vs Infant Mortality Rate

 - **Infant mortality rate** : number of infant deaths before his or her first birthday for every 1000 live births
 - **Population using improved sanitation facilities** : percentage of people in a country that can enjoy this facilities 
 - **Population using improved drinking water** : percentage of people in a country that can enjoy this facilities 

```python
#Sanitation and drinking water vs infant mortality rate
drink_san_mortality =pd.DataFrame(data[['Infant mortality rate (per 1000 live births',
                                        'Pop. using improved sanitation facilities (rural, %)',
                                        'Pop. using improved sanitation facilities (urban, %)',
                                        'Pop. using improved drinking water (rural, %)',
                                        'Pop. using improved drinking water (urban, %)']])
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(5, 3, figsize=(30, 22))
colors = ["r", "g", "b", "m", "c", "y"]
count = 0
for var in drink_san_mortality:
    sb.boxplot(data = drink_san_mortality[var], orient = "h", ax = axes[count,0], color = colors[count])
    sb.histplot(data = drink_san_mortality[var], ax = axes[count,1], color = colors[count])
    sb.violinplot(data = drink_san_mortality[var], orient = "h", ax = axes[count,2], color = colors[count])
    count += 1
```

### Analysis
- Most countries have low infant mortality rates (under 10 deaths per 1000 live births)
- Trends in sanitation and drinking water facilities in rural and urban areas are similar.
- Most countries have upgraded both of these facilities (above 90%).
- From these two observations, it can be concluded that there is a negative relationship between infant mortality rate and improvement in sanitation & drinking water facilities.
- The better the facilities, the lower the infant mortality rate.

```python
# Calculate the correlation between variables
sb.heatmap(drink_san_mortality.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```


The **dark red colors** indicates there is a strong negative correlation of each independent variable with infant mortality rate

```python
# Representation of relationship with jointplot (sanitation)
sb.jointplot(data = drink_san_mortality, x = "Pop. using improved sanitation facilities (rural, %)", y = "Infant mortality rate (per 1000 live births", height = 5)
sb.jointplot(data = drink_san_mortality, x = "Pop. using improved sanitation facilities (urban, %)", y = "Infant mortality rate (per 1000 live births", height = 5)
```

```python
# Representation of relationship with jointplot (drinking water)
sb.jointplot(data = drink_san_mortality, x = "Pop. using improved drinking water (rural, %)", y = "Infant mortality rate (per 1000 live births", height = 5)
sb.jointplot(data = drink_san_mortality, x = "Pop. using improved drinking water (urban, %)", y = "Infant mortality rate (per 1000 live births", height = 5)
```

## Region vs Happiness score

```python
print ("Number of Region: ", len (data['Region'].unique()))

print(data['Region'].value_counts())
sb.catplot(y='Region', data = data, kind = 'count')
```

```python
# Region vs Happiness score
f = plt.figure(figsize=(16, 8))
sb.stripplot(x = 'happiness.score', y = 'Region', data = data)
```

```python
# Group Happiness score by the categorical variable to find the medians for each boxplot above and sort accordingly
f = plt.figure(figsize=(16, 8))
sb.boxplot(x = 'happiness.score', y = 'Region', data = data, 
           order = data.groupby('Region')['happiness.score'].median().sort_values().index)
```

- Most countries in African Region have a happiness value below five.
- While countires in American Region have a happiness value above five. 
- Countries in Western Asia has the most dispersed happiness.score, eg: some have very low scores others have very high scores. 

# Happiness score vs Life expectancy (male and female)

```python
life_expectancy =pd.DataFrame(data[['Life expectancy at birth (females, years)',
                                   'Life expectancy at birth (males, years)',
                                   'happiness.score']])
# Draw the distributions of all variables
f, axes = plt.subplots(3, 3, figsize=(30, 22))
colors = ["m", "c", "y"]
count = 0
for var in life_expectancy:
    sb.boxplot(data = life_expectancy[var], orient = "h", ax = axes[count,0], color = colors[count])
    sb.histplot(data = life_expectancy[var], ax = axes[count,1], color = colors[count])
    sb.violinplot(data = life_expectancy[var], orient = "h", ax = axes[count,2], color = colors[count])
    count += 1
```

- Majority countries' happiness scores range from 4 to 6.
- Majority of females in the world have life expectancy between 75- 80 years 
- Majority of males in the world have life expectancy between 70-80 years. 


```python
sb.heatmap(life_expectancy.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```

There is a storng positive correlation between happiness score and life expectancy (male and female)

```python
sb.jointplot(data = life_expectancy, x = "Life expectancy at birth (females, years)", y = "happiness.score", height = 5)
sb.jointplot(data = life_expectancy, x = "Life expectancy at birth (males, years)", y = "happiness.score", height = 5)
```

```python
# Group Life expectancy (males) by the categorical variable to find the medians for each boxplot above and sort accordingly
f = plt.figure(figsize=(16, 8))
sb.boxplot(x = 'Life expectancy at birth (males, years)', y = 'Region', data = data, 
           order = data.groupby('Region')['Life expectancy at birth (males, years)'].median().sort_values().index)
```

```python
# Group Life expectancy (males) by the categorical variable to find the medians for each boxplot above and sort accordingly
f = plt.figure(figsize=(16, 8))
sb.boxplot(x = 'Life expectancy at birth (females, years)', y = 'Region', data = data, 
           order = data.groupby('Region')['Life expectancy at birth (females, years)'].median().sort_values().index)
```

- Middle, Western, and Eastern Africa has the lowest life expectancy at birth for both female and male.
- For male it is below 65 years and for female it is below 70 years old.
<br>
- Most countries in western Europe and Oceania has the longest life expectancy.
