### Insight 1. Does a population with highly educated females lead to lower birth rates?
> Extract required variables from 'Data' dataframe for analysis  
> - `Education: Primary gross enrol. ratio (female per 100 pop.)`   
> - `Education: Secondary gross enrol. ratio (female per 100 pop.)`  
> - `Education: Tertiary gross enrol. ratio (female per 100 pop.)`  
> - `Fertility rate, total (live births per woman)`

```python
EduLvl_BirthRate_Data = pd.DataFrame(Data[['Education: Primary gross enrol. ratio (female per 100 pop.)', 
                                           'Education: Secondary gross enrol. ratio (female per 100 pop.)', 
                                           'Education: Tertiary gross enrol. ratio (female per 100 pop.)', 
                                           'Fertility rate, total (live births per woman)']])
EduLvl_BirthRate_Data.head()
```

### Check variables independently

```python
EduLvl_BirthRate_Data.describe()
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(4, 3, figsize=(18, 20))

count = 0
for var in EduLvl_BirthRate_Data:
    sb.boxplot(data = EduLvl_BirthRate_Data[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = EduLvl_BirthRate_Data[var], ax = axes[count,1])
    sb.violinplot(data = EduLvl_BirthRate_Data[var], orient = "h", ax = axes[count,2])
    count += 1
```

```python
# Calculate the quartiles
Q1 = EduLvl_BirthRate_Data.quantile(0.25)
Q3 = EduLvl_BirthRate_Data.quantile(0.75)

# Rule to identify outliers
rule = ((EduLvl_BirthRate_Data < (Q1 - 1.5 * (Q3 - Q1))) | (EduLvl_BirthRate_Data > (Q3 + 1.5 * (Q3 - Q1))))

# Count the number of outliers
rule.sum()
```

```python
EduLvl_BirthRate_Data.skew()
```

By observation and computation, we can tell that `Education: Primary gross enrol. ratio (female per 100 pop.)` has the most number of outliers compared to the other variables while `Fertility rate, total (live births per woman)` is moderately skewed at 1.042320.

### Check relationship between education levels and fertility rate

```python
EduLvl_BirthRate_Data[EduLvl_BirthRate_Data.columns[:]].corr()['Fertility rate, total (live births per woman)'][:].sort_values()
```

```python
# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(12, 12))
sb.heatmap(EduLvl_BirthRate_Data.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = EduLvl_BirthRate_Data, height=5)
```

```python
cols = list(EduLvl_BirthRate_Data.columns)[:-1]
# Draw the distributions of all variables
count = 0
for item in cols:
    g = sb.jointplot(x = item, 
                 y = "Fertility rate, total (live births per woman)", 
                 data = EduLvl_BirthRate_Data,
                 height = 12,
                 ax = axes[0], 
                 kind = "reg")
    j = sb.regplot(x=item, y="Fertility rate, total (live births per woman)", 
               data=EduLvl_BirthRate_Data, ax = g.ax_joint)
    j.set(ylim=(0, None))
    count += 1
```

- Compared to secondary and tertiary education level, primary education level has the lowest correlation value compared to fertility rate.  

- It is difficult to derive any conclusion from the primary education level joint plot as most data points are centralised at about 100 (female per 100 in the population) and 1 - 3 fertility rate.  

- However, we can tell that between the three education levels and the fertility rate, all three are negatively correlated. Especially in secondary and tertiary education levels, it is clear that as education enrolment ratio increases, the fertility rate decreases.  

- This allows us to make an observation that birth rate decreases as the female population gets more educated, but, this is not conclusive as there may be other factors affecting the fertility rate.

### Insight 2. Happiness Score Compared to Locality
> Extract required variables from 'Data' dataframe for analysis  
> - `country`   
> - `region`  
> - `happiness.score`  
> - `latitude`  
> - `longitude`

```python
Map_Data = pd.DataFrame(Data[['country', 'Region', 'happiness.score', 'latitude', 'longitude']])
Map_Data.head()
```

```python
Map_Data.info()
```

```python
import geopandas as gpd

# From GeoPandas, our world map data
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Creating axes and plotting world map
fig, ax = plt.subplots(figsize=(20, 10))
worldmap.plot(color="lightgrey", ax=ax)

# Plotting our happiness score with a color map
x = Map_Data['longitude']
y = Map_Data['latitude']
z = Map_Data['happiness.score']
plt.scatter(x, y, s=20*z, c=z, alpha=0.6, cmap='RdBu')
plt.colorbar(label='happiness.score')

# Creating axis limits and title
plt.xlim([-180, 180])
plt.ylim([-90, 90])

plt.title("World Happiness Report\n")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

- We can observe that the regions experiencing the lowest happiness scores are located primarily in Africa, the Middle East and South Asia.
- On the flip side, the scandinavian countries are experiencing the highest happiness scores compared to the rest of the world.
- It seems like a country's happiness score can be influenced by the neighbouring countries. However, an anomaly can be seen in countries like Qatar, United Arab Emirates and Saudi Arabia, who experience higher happiness scores compared to the countries near it in Africa and South Asia.

```python
print ("Number of Region: ", len (Map_Data['Region'].unique()))

print(Map_Data['Region'].value_counts())
sb.catplot(y='Region', data = Map_Data, kind = 'count')
```

```python
# Region vs Happiness score
f = plt.figure(figsize=(16, 8))
sb.stripplot(x = 'happiness.score', y = 'Region', data = Map_Data)
```

```python
# Group Happiness score by the categorical variable to find the medians for each boxplot above and sort accordingly
f = plt.figure(figsize=(16, 8))
sb.boxplot(x = 'happiness.score', y = 'Region', data = Map_Data, 
           order = Map_Data.groupby('Region')['happiness.score'].median().sort_values().index)
```

- Most countries in African Region have a happiness value below five.
- While countires in American Region have a happiness value above five. 
- Countries in Western Asia has the most dispersed happiness.score, eg: some have very low scores others have very high scores.

### Check variables independently

```python
drink_san_mortality.describe()
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(5, 3, figsize=(30, 22))
colors = ["r", "g", "b", "m", "c", "y"]
count = 0
for var in drink_san_mortality:
    sb.boxplot(data = drink_san_mortality[var], orient = "h", ax = axes[count,0], color = colors[count])
    sb.histplot(data = drink_san_mortality[var], ax = axes[count,1], color = colors[count])
    sb.violinplot(data = drink_san_mortality[var], orient = "h", ax = axes[count,2], color = colors[count])
    count += 1
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = drink_san_mortality, 
            x_vars = ['Pop. using improved sanitation facilities (rural, %)', 
                      'Pop. using improved sanitation facilities (urban, %)', 
                      'Pop. using improved drinking water (rural, %)', 
                      'Pop. using improved drinking water (urban, %)'], 
            y_vars=['Infant mortality rate (per 1000 live births'], height=4)
```

- Most countries have low infant mortality rates (under 10 deaths per 1000 live births)
- Trends in sanitation and drinking water facilities in rural and urban areas are similar.
- Most countries have upgraded both of these facilities (above 90%).
- From these two observations, it can be concluded that there is a negative relationship between infant mortality rate and improvement in sanitation & drinking water facilities.
- The better the facilities, the lower the infant mortality rate.

```python
# Calculate the quartiles
Q1 = employment_df.quantile(0.25)
Q3 = employment_df.quantile(0.75)

# Rule to identify outliers
rule = ((employment_df < (Q1 - 1.5 * (Q3 - Q1))) | (employment_df > (Q3 + 1.5 * (Q3 - Q1))))

# Count the number of outliers
rule.sum()
```

```python
f = plt.figure(figsize=(12, 12))
sb.heatmap(employment_df.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
agri = 'Employment: Agriculture (% of employed)'
serv = 'Employment: Services (% of employed)'
indu = 'Employment: Industry (% of employed)'
net = 'Individuals using the Internet (per 100 inhabitants)'

plt.figure(figsize=(15,15))
plt.scatter(employment_df[agri], employment_df[net], s = 100, label = 'Agriculture')
plt.scatter(employment_df[serv], employment_df[net], s = 100, label = 'Services')
plt.scatter(employment_df[indu], employment_df[net], s = 100, label = 'Industry')

plt.legend(loc='upper right', frameon=True)
plt.xlabel('Employment: (% of employed)')
plt.ylabel('Individuals using the Internet (per 100 inhabitants)')

plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.scatter(employment_df[agri], employment_df[net], s = 100, label = 'Agriculture')
plt.scatter(employment_df[serv], employment_df[net], s = 100, label = 'Services')
ax = sb.regplot(x=agri, y=net, data=employment_df)
ax = sb.regplot(x=serv, y=net, data=employment_df)
ax.set(ylim=(0, None))

plt.legend(loc='upper right', frameon=True)
plt.xlabel('Employment: (% of employed)')
plt.ylabel('Individuals using the Internet (per 100 inhabitants)')
```

- As one can see there is a trend line of a relation between certain employment against internet usage.  
- Lesser individuals are using the internet if a higher percentage of them are employed in agriculture, while more individuals are using the internet if a higher percentage of them are employed in services.
- Perhaps due to the nature of services, it is required to have internet usage for services to be done better.  
- In contrast to agriculture where more time would be spend on the field leading to lesser time on the internet
- Due to the weak correlation between internet usage and industrial employment, it is difficult to make an observation.
